# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

## Loading Categorized Image Data from a `.npz` File
- The function `np.load` is used to load the contents of the `.npz` file, which contains the image data and its corresponding labels and sub-labels.
- The `allow_pickle=True` argument is used to allow loading objects that may have been saved using Python's pickle format, which is necessary for arrays containing non-basic data types.
 
**Loaded Data**:
   - `image_pixels`: The array containing the preprocessed image pixel data.
   - `labels`: The array containing the labels for the images (e.g., `0` for good images, `1` for manipulated images).
   - `sub_labels`: The array containing the sub-labels for the images (e.g., "good", "manipulated_front").

```python
# Load the saved .npz file containing image data, labels, and sub-labels
data = np.load('data.npz', allow_pickle=True)
image_pixels = data['image']
labels = data['label']
sub_labels = data['sub_label']
```

# Exploring the Data

```python
print("Image Pixel Data Shape:", image_pixels.shape)
```

## Extracting Unique Labels and Sub-Labels with Counts

```python
# Get the unique labels and their respective counts
unique_labels, label_counts = np.unique(labels, return_counts=True)

# Get the unique sub-labels and their respective counts
unique_sub_labels, sub_label_counts = np.unique(sub_labels, return_counts=True)
```

```python
# Plot the distribution of labels
plt.figure(figsize=(10, 5))
plt.bar(unique_labels, label_counts)
plt.xlabel('Labels')
plt.ylabel('Frequency')
plt.title('Distribution of Labels')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(unique_sub_labels, sub_label_counts, color='skyblue')
plt.xlabel('Sub-labels')
plt.ylabel('Frequency')
plt.title('Distribution of Sub-labels')
plt.show()
```

## Identifying the Minority Label in the Dataset

```python
# Count the occurrences of each label
label_counts = Counter(labels)

# Find the minority class (the label with the minimum count)
minority_class = min(label_counts, key=label_counts.get)

# Display the minority class
minority_class
```

## Image Augmentation for Minority Class

Image augmentation is performed on the minority class to balance the dataset. The `ImageDataGenerator` from Keras is used to apply random transformations such as rotation and horizontal flipping to the images of the minority class. The augmented images are then stored along with their corresponding labels and sub-labels.

```python
datagen = ImageDataGenerator(
    rotation_range=30,
    horizontal_flip=True,
)
```

```python
augmented_images = []
augmented_labels = []
augmented_sub_labels = []
```

```python
# Augment the images from the minority class
for img, sub_label in zip(image_pixels[labels == minority_class], sub_labels[labels == minority_class]):
    img = img.reshape((1,) + img.shape)  # Reshape for ImageDataGenerator
    # Generate augmented images and add them to the lists
    for i, batch in enumerate(datagen.flow(img, batch_size=1)):
        augmented_images.append(batch[0])
        augmented_labels.append(minority_class)
        augmented_sub_labels.append(sub_label)  # Add the sub-label for the augmented image
        if i >= 2:  # Limit to 3 augmented images per original image
            break
```

```python
# Concatenate the original and augmented data to create a balanced dataset
balanced_images = np.concatenate((image_pixels, augmented_images), axis=0)
balanced_labels = np.concatenate((labels, augmented_labels), axis=0)
balanced_sub_labels = np.concatenate((sub_labels, augmented_sub_labels), axis=0)
```

## Checking if Dataset is now balanced or not

```python
unique_labels, label_counts = np.unique(balanced_labels, return_counts=True)
```

```python
# Plot the distribution of labels
plt.figure(figsize=(10, 5))
plt.bar(unique_labels, label_counts)
plt.xlabel('Labels')
plt.ylabel('Frequency')
plt.title('Distribution of Labels')
plt.show()
```

## Saving Balanced Dataset as a `.npz` File

```python
# Save as .npz file
np.savez('balanced_data.npz', 
         image=balanced_images, 
         label=balanced_labels,
        sub_label=balanced_sub_labels)
```
